```python
from model import kNN
import numpy as np
from matplotlib import pyplot as plt

X = np.array(
    [[np.random.normal(1, 0.6), np.random.normal(1, 0.6)] for _ in range(50)] +
    [[np.random.normal(2, 0.6), np.random.normal(2, 2.2)] for _ in range(50)]
)

y = np.array(
    [0 for _ in range(50)] +
    [1 for _ in range(50)]
)
```

```python
model = kNN(7, n_trees=10, max_leaf_elems=7)
model.fit(X, y)
print(model.predict(np.array([2.5, 2.0])))

fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(12, 12))
plt.axes(axs)
xlim = [np.min(X[:, 0]) - 1, np.max(X[:, 0]) + 1]
ylim = [np.min(X[:, 1]) - 1, np.max(X[:, 1]) + 1]
xx, yy = np.meshgrid(np.linspace(*xlim, num=400), np.linspace(*ylim, num=400))
arr = np.c_[xx.ravel(), yy.ravel()]
ys = np.empty(shape=(arr.shape[0]))
for i in range(len(arr)):
    ys[i] = model.predict(arr[i])
    if i % 100 == 0:
        print(i)
z_model = ys.reshape(xx.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='bwr')
plt.contour(xx, yy, z_model, colors='k', levels=[-1, 0, 1], alpha=0.75, linestyles=['--', '-', '--'])
plt.contourf(xx, yy, np.sign(z_model.reshape(xx.shape)), alpha=0.3, levels=2, cmap='coolwarm', zorder=1)
```
